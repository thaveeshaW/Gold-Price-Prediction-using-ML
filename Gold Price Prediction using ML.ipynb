Import Dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics
```

Collect Data & Process

```python
# Load the csv data to a Pandas DataFrame
gold_data = pd.read_csv('/content/gld_price_data.csv')
```

```python
# Print first 5 rows in the dataframe
gold_data.head()
```

```python
# Print last 5 rows of the dataframe
gold_data.tail()
```

```python
# Number of rows and columns
gold_data.shape
```

```python
# Get basic information about the data
gold_data.info()
```

```python
# Check the number of missing values
gold_data.isnull().sum()
```

```python
# Get the statistical measures of the data
gold_data.describe()
```

```python
# Find Correlation
correlation = gold_data.drop(['Date'],axis=1).corr()
```

```python
# Constructin a heatmap to understand the correlatiom
plt.figure(figsize = (8,8))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f',annot=True, annot_kws={'size':8}, cmap='Reds')
```

```python
# Correlation values of GLD
print(correlation['GLD'])
```

```python
# Check the distribution of  GLD Price
sns.displot(gold_data['GLD'],color='blue')
```

Split the Features and Targets

```python
X = gold_data.drop(['Date','GLD'],axis=1)
Y = gold_data['GLD']
```

```python
print(X)
```

```python
print(Y)
```

Split into Training Data and Test Data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state=2)
```

Train the Model: Random Forest Regressor

```python
regressor = RandomForestRegressor(n_estimators=100)
```

```python
# Train the model
regressor.fit(X_train,Y_train)
```

Model Evaluation

```python
# Prediction on Test Data
test_data_prediction = regressor.predict(X_test)
```

```python
print(test_data_prediction)
```

```python
# R Squared Error
error_score = metrics.r2_score(Y_test, test_data_prediction)
print("R Squared error : ", error_score)
```

Plot Actual prices and Predicted prices for Comparison

```python
Y_test = list(Y_test)
```

```python
plt.plot(Y_test, color='black', label = 'Actual Value')
plt.plot(test_data_prediction, color='red', label='Predicted Value')
plt.title('Actual Price vs Predicted Price')
plt.xlabel('No of values')
plt.ylabel('GLD Price')
plt.legend()
plt.show()
```
